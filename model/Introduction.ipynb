```python
data=pd.read_csv('accalldata.csv')
data.head()
```

```python
data['subset_region']=data['region'].str.split(expand=True)[2]
```

```python
names={'time':'시간','daynight':'주야구분','dayofweek':'요일','lon':'위도','lat':'경도','die':'사망자수',
       'injure':'부상자수','injure_mid':'중상자수','injure_sm':'경상자수','acc_bg':'사고유형대분류','acc_mid':'사고유형중분류',
      'acc_sm':'사고유형소분류','region':'주소','subset_region':'지역'}

data=data.rename(columns=names)
```

```python
data['시간']=pd.to_datetime(data['시간'],format='%Y%m%d%H')

data['연']=data['시간'].dt.year
data['월']=data['시간'].dt.month
data['일']=data['시간'].dt.day
data['시간']=data['시간'].dt.hour
```

```python
import os 

temps=['./temp/'+d for d in os.listdir('./temp')]
```

```python
temp_2012=pd.read_csv(temps[0],encoding='cp949')
temp_2013=pd.read_csv(temps[1],encoding='cp949')
temp_2014=pd.read_csv(temps[2],encoding='cp949')
temp_2015=pd.read_csv(temps[3],encoding='cp949')
temp_2016=pd.read_csv(temps[4],encoding='cp949')
temp_2017=pd.read_csv(temps[5],encoding='cp949')
temp_2018=pd.read_csv(temps[6],encoding='cp949')
temp_2019=pd.read_csv(temps[7],encoding='cp949')
temp_2020=pd.read_csv(temps[8],encoding='cp949')
```

```python
def conv_merge(df1,df2):
    df1['일시']=pd.to_datetime(df1['일시'],format='%Y-%m-%d %H:%M')
    
    df1['연']=df1['일시'].dt.year
    df1['월']=df1['일시'].dt.month
    df1['일']=df1['일시'].dt.day
    df1['시간']=df1['일시'].dt.hour
    
    df1=df1[['연','월','일','시간','기온(°C)','풍향(deg)','풍속(m/s)','강수량(mm)','습도(%)']]
    df=pd.merge(df1,df2,on=['연','월','일','시간'],how='right')
    return df
```

```python
df1=conv_merge(temp_2012,data).dropna()
df2=conv_merge(temp_2013,data).dropna()
df3=conv_merge(temp_2014,data).dropna()
df4=conv_merge(temp_2015,data).dropna()
df5=conv_merge(temp_2016,data).dropna()
df6=conv_merge(temp_2017,data).dropna()
df7=conv_merge(temp_2018,data).dropna()
df8=conv_merge(temp_2019,data).dropna()
df9=conv_merge(temp_2020,data).dropna()
```

```python
data=pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9])
```

```python
data.to_csv('accalldata.csv',index=False)
```
